# I. Définir un objectif d'étude mesurable

Objectif : Se fixer un objectif pour notre étude, des metriques d'évaluation de la performance du modèle, et des valuers visées pour ces métriques

les métriques souvent utilisées sont F1, Recall, Précision
$$\text{Précision} = \frac{\text{Vrai-Positif}}{\text{Vrai-Positif} + \text{Faux-Positif}}$$
$$\text{Recall} = \frac{\text{Vrai-Positif}}{\text{Vrai-Positif} + \text{Faux-Négatif}}$$
$$\text{F1} = \frac{\text{Précision}}{\text{Recall}}$$
